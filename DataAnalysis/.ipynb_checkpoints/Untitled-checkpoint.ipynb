```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('epl.csv')

df
```

```python
home_points = []
away_points = []
for i in range(1, len(df.home_team_goal_count)):
    if df.home_team_goal_count[i] > df.away_team_goal_count[i]:
        home_points.append(3)
        away_points.append(0)
    elif df.home_team_goal_count[i] < df.away_team_goal_count[i]:
        home_points.append(0)
        away_points.append(3)
    else:
        home_points.append(1)
        away_points.append(1)
        
df2 = pd.DataFrame (home_points, columns = ['home_pts'])

df1 = pd.concat([df,df2])

df2
```
